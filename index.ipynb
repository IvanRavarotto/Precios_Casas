# Predicción de precios de casas (Supervisado - Regresión)
* Variables: metros_cuadrados, habitaciones, baños, ubicación, antigüedad
* Objetivo: precio
* Tarea: Entrenar un modelo para predecir el precio de una casa.

### Importación de librerias y creación de variables

```python
# Librerias importadas + instalación mediante pip install:
#pip install pandas
import pandas as pd
#pip install matplotlib
import matplotlib.pyplot as plt
#pip install scikit-learn
from sklearn.model_selection import train_test_split



# Carga de datos y creacion de variables:
df = pd.read_csv('dataset_precios_casas.csv')
Y = df['ubicacion']
X = df.drop(columns=['ubicacion'])
```

### Unicos, minimos, maximos, medio y medianos | Separación entre datos para entrenar y prueba

```python
#Valores unicos por columna:
valoresUnicos = X.nunique()
print("Valores unicos por columna: ", valoresUnicos)
```

```python
#Valores Minimos y Maximos:
valoresMinimos = X.min()
print("Valores Minimos por columna: ", valoresMinimos)
valoresMaximos = X.max()
print("Valores Maximos por columna: ", valoresMaximos)
```

```python
#Medianos y medios por columna:
valoresMedianos = X.mean()
print("Valores Medianos por columna: ", valoresMedianos)
valoresMedios = X.median()
print("Valores Medios por columna: ", valoresMedios)
```

```python
# Separación de datos para entrenar y probar el modelo (70% - 30%):
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
#train_test_split divide los datos en conjuntos de entrenamiento y prueba
#test_size indica el porcentaje de datos que se usarán para pruebas
#random_state se usa para asegurar que la división sea reproducible

print("Datos de entrenamiento (X_train):\n", X_train)
print("Datos de prueba (X_test):\n", X_test)
```
