* Tarea: Entrenar un modelo para predecir el precio de una casa.

### Importación de librerias y creación de variables:

```python
# Librerias importadas:
import pandas as pd
import matplotlib.pyplot as plt

# Carga de datos y creacion de variables:
df = pd.read_csv('dataset_precios_casas.csv')
Y = df['ubicacion']
X = df.drop(columns=['ubicacion'])
```

### Unicos, minimos, maximos, medio y medianos

```python
#Valores unicos por columna:
valoresUnicos = X.nunique()
print("Valores unicos por columna: ", valoresUnicos)
```

```python
#Valores Minimos y Maximos:
valoresMinimos = X.min()
print("Valores Minimos por columna: ", valoresMinimos)
valoresMaximos = X.max()
print("Valores Maximos por columna: ", valoresMaximos)
```

```python
#Medianos y medios por columna:
valoresMedianos = X.mean()
print("Valores Medianos por columna: ", valoresMedianos)
valoresMedios = X.median()
print("Valores Medios por columna: ", valoresMedios)
```
